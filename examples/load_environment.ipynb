```python
%matplotlib inline
import matplotlib.pyplot as plt
import pybullet as p
import pybullet_data

p.connect(p.GUI, options='--background_color_red=1.0 --background_color_green=1.0 --background_color_blue=1.0');
```

```python
for Env in [SnakeEnv, UR5Env, KukaEnv, ExtendKukaEnv]:
    env = Env(objects=[VoxelObject(base_orientation=[0, 0, 0, 1], base_position=[1, 1, 1], half_extents=[0.2, 0.2, 0.2])])
    env.load(reconnect=False, GUI=True, light_height_z=0.1)
    plt.clf()
    plt.close('all')
    plt.imshow(env.render())
    plt.show()
```
